```python
loanAmount = float(input("Loan amount"))
numYears = int(input("Number of  years"))
interestRate = float(input("Interest rate"))
monthlyInterestRate = interestRate / 1200

denominator = 1 - (1 / (1 + monthlyInterestRate) ** (numYears * 12))
numerator = loanAmount * monthlyInterestRate
monthlyPay = numerator / denominator
totalPay = monthlyPay * numYears * 12

balance = loanAmount

print(monthlyPay)
print(totalPay)

print("Payment#\t\tInterest\t\tPrincipal\t\tBalance")
for i in range(1, numYears * 12 + 1):
    interest = monthlyInterestRate * balance
    principal = monthlyPay - interest
    balance = balance - principal
    if balance <= 0:
        print("you're done paying")
        break;
    print(i, "\t\t", '{:.{}f}'.format(interest, 2), "\t\t", '{:.{}f}'.format(principal, 2), "\t\t\t",
        '{:.{}f}'.format(balance, 2))

```
