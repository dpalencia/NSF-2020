```python
#Import some Pandas and Numpy
import pandas as pd
import numpy as np
```

```python
#Read a file as a dataframe and figure out the columns.
cityAtts = pd.read_csv("Weather Data/city_attributes.csv")

#See what data we have.
print(cityAtts.shape)
cityAtts.head()
```

```python
#Take a look at rows with any null value.
#Get a boolean array. True when any item across the row is null
bools = cityAtts.isnull().any(axis=1)
cityAtts.loc[bools]
```

```python
#Drop rows with null values.
cityAtts.dropna()
```

```python
#Convert a colummn to  a numpy floating point array.
#Get the latitudes as a list and turn it into a numpy array.
latFloats = np.array(cityAtts.loc[:,"Latitude"])
print(latFloats)
```

```python
#Compute the mean of this column.
latMean = np.mean(latFloats)
print(latMean)
```

```python
#Commpute the std of this column.
latStd = np.std(latFloats)
print(latStd)
```

```python
#Compute correlation of two columns.
longFloats = cityAtts.loc[:,"Longitude"]
latCorr = np.corrcoef(latFloats, longFloats)
print(latCorr)
```

```python
#Read the temperature data. See shape and head
tempFrame = pd.read_csv("Weather Data/temperature.csv")
print(tempFrame.shape)
print(tempFrame.columns)
tempFrame.head()
```

```python
#Calculate the average temperatures and add this as a new column in our city dataframe.
#Check if the column already exists to avoid python errors
avgTemps = tempFrame.mean()
print(avgTemps.values)
```

```python
#Insert a new average temps column in our city dataframe so we can do some plotting
if 'avgTemp' not in cityAtts.columns:
    cityAtts.insert(loc=4, value=avgTemps.values, column="avgTemp")
cityAtts.loc[:,"avgTemp"] = avgTemps.values
cityAtts.head()
```

```python
#Now plot latitude against average temperature
import matplotlib  as plt
plt.ylabel = "Average Temperature"
cityAtts.plot(x = "Latitude", y = "avgTemp", kind = "scatter")
#neat.
```
